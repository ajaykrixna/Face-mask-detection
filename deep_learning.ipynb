```python
# ================================================================
# TRAIN: FACE MASK DETECTION USING CNN (GPU-Optimized for P100)
# ================================================================
import os, json, numpy as np, tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPool2D, Flatten, Dense, Dropout, BatchNormalization, Input
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt

# ----------------------------
# GPU CHECK & CONFIG
# ----------------------------
print("Num GPUs Available:", len(tf.config.list_physical_devices('GPU')))
gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
        print("✅ GPU Memory growth enabled")
    except RuntimeError as e:
        print(e)

# Enable mixed precision for speed on P100
from tensorflow.keras import mixed_precision
mixed_precision.set_global_policy('mixed_float16')
print("✅ Mixed precision policy:", mixed_precision.global_policy())

# ----------------------------
# Paths & hyperparameters
# ----------------------------
ROOT = "/kaggle/input/face-mask-dataset/data"
WORK = "/kaggle/working"
MODEL_SAVE = os.path.join(WORK, "mask_detector_cnn_best.h5")
LABEL_MAP_JSON = os.path.join(WORK, "label_map.json")

IMG_SIZE = (128, 128)
BATCH = 32
EPOCHS = 30
LR = 1e-3
SEED = 42

# ----------------------------
# Data generators: enforce class order
# ----------------------------
train_datagen = ImageDataGenerator(
    rescale=1./255.,
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    validation_split=0.2
)
val_datagen = ImageDataGenerator(rescale=1./255., validation_split=0.2)

# IMPORTANT: specify classes order so 0 -> with_mask, 1 -> without_mask
CLASS_ORDER = ["with_mask", "without_mask"]

train_gen = train_datagen.flow_from_directory(
    ROOT,
    target_size=IMG_SIZE,
    batch_size=BATCH,
    class_mode="binary",
    subset="training",
    seed=SEED,
    shuffle=True,
    classes=CLASS_ORDER
)
val_gen = val_datagen.flow_from_directory(
    ROOT,
    target_size=IMG_SIZE,
    batch_size=BATCH,
    class_mode="binary",
    subset="validation",
    seed=SEED,
    shuffle=False,
    classes=CLASS_ORDER
)

# ----------------------------
# Save label map index -> label
# ----------------------------
# train_gen.class_indices maps {'with_mask':0, 'without_mask':1}
label_map = {v: k for k, v in train_gen.class_indices.items()}  # {0: 'with_mask', 1: 'without_mask'}
with open(LABEL_MAP_JSON, "w") as f:
    json.dump(label_map, f)
print("✅ Saved label_map (index->label):", label_map)

# ----------------------------
# Class weights (handle imbalance)
# ----------------------------
y = train_gen.classes
classes = np.unique(y)
class_weights = compute_class_weight(class_weight='balanced', classes=classes, y=y)
class_weight = {int(c): float(w) for c, w in zip(classes, class_weights)}
print("✅ Using class_weight:", class_weight)

# ----------------------------
# Build CNN (final Dense dtype float32 to avoid float16 output issues)
# ----------------------------
def build_cnn(input_shape=IMG_SIZE + (3,)):
    model = Sequential([
        Input(shape=input_shape),
        Conv2D(32, (3,3), activation='relu', padding='same'),
        BatchNormalization(),
        MaxPool2D(),
        Dropout(0.25),

        Conv2D(64, (3,3), activation='relu', padding='same'),
        BatchNormalization(),
        MaxPool2D(),
        Dropout(0.25),

        Conv2D(128, (3,3), activation='relu', padding='same'),
        BatchNormalization(),
        MaxPool2D(),
        Dropout(0.3),

        Flatten(),
        Dense(128, activation='relu'),
        BatchNormalization(),
        Dropout(0.4),
        Dense(1, activation='sigmoid', dtype='float32')  # ensure float32 output
    ])
    return model

model = build_cnn()
# Use LossScaleOptimizer when mixed precision is enabled
base_opt = Adam(learning_rate=LR)
try:
    opt = mixed_precision.LossScaleOptimizer(base_opt)
    print("✅ Wrapped optimizer with LossScaleOptimizer for mixed precision")
except Exception:
    opt = base_opt

model.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])
model.summary()

# ----------------------------
# Callbacks
# ----------------------------
callbacks = [
    ModelCheckpoint(MODEL_SAVE, monitor='val_accuracy', save_best_only=True, verbose=1),
    EarlyStopping(monitor='val_loss', patience=6, restore_best_weights=True, verbose=1)
]

# ----------------------------
# Train
# ----------------------------
history = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=EPOCHS,
    callbacks=callbacks,
    class_weight=class_weight,
    verbose=1
)

# ----------------------------
# Evaluate on validation
# ----------------------------
val_gen.reset()
y_true = val_gen.classes
y_prob = model.predict(val_gen, verbose=1).ravel()
y_pred = (y_prob > 0.5).astype(int)

val_accuracy = accuracy_score(y_true, y_pred)
print(f"\n✅ Validation Accuracy: {val_accuracy * 100:.2f}%")

# target names in index order
target_names = [label_map[i] for i in sorted(label_map.keys())]
print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=target_names))
print("Confusion Matrix:")
print(confusion_matrix(y_true, y_pred))

# ----------------------------
# Plot training curves
# ----------------------------
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(history.history['loss'], label='train_loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.legend(); plt.title("Loss")
plt.subplot(1,2,2)
plt.plot(history.history['accuracy'], label='train_acc')
plt.plot(history.history['val_accuracy'], label='val_acc')
plt.legend(); plt.title("Accuracy")
plt.show()

print("\n✅ Best model saved to:", MODEL_SAVE)
print("✅ Label map saved to:", LABEL_MAP_JSON)
print("✅ Training complete.")
```

```python
# Robust inference using DNN SSD -> Haar -> Sliding-window
import os, json, cv2, numpy as np, tensorflow as tf
from tensorflow.keras.models import load_model
from matplotlib import pyplot as plt

MODEL_PATH = "/kaggle/working/mask_detector_cnn_best.h5"
LABEL_MAP = "/kaggle/working/label_map.json"
WORK = "/kaggle/working"
PROTO = os.path.join(WORK, "deploy.prototxt")
CAFFE = os.path.join(WORK, "res10_300x300_ssd_iter_140000.caffemodel")
IMG_SIZE = (128,128)

# load model & labels
model = load_model(MODEL_PATH)
with open(LABEL_MAP,"r") as f:
    idx2label = json.load(f)
idx2label = {int(k):v for k,v in idx2label.items()}

# prepare detectors
dnn_net = None
if os.path.exists(PROTO) and os.path.exists(CAFFE):
    try:
        dnn_net = cv2.dnn.readNetFromCaffe(PROTO, CAFFE)
        print("Using OpenCV DNN SSD detector.")
    except Exception as e:
        print("Failed to load DNN SSD:", e)

haar = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
if haar.empty():
    raise RuntimeError("Haar cascade not available")

# helpers
def detect_dnn(img_bgr, net, conf_thresh=0.5):
    h,w = img_bgr.shape[:2]
    blob = cv2.dnn.blobFromImage(cv2.resize(img_bgr,(300,300)), 1.0, (300,300), (104.0,177.0,123.0))
    net.setInput(blob)
    det = net.forward()
    boxes=[]
    for i in range(det.shape[2]):
        conf = float(det[0,0,i,2])
        if conf>conf_thresh:
            x1=int(det[0,0,i,3]*w); y1=int(det[0,0,i,4]*h)
            x2=int(det[0,0,i,5]*w); y2=int(det[0,0,i,6]*h)
            boxes.append((max(0,x1), max(0,y1), max(1,x2-x1), max(1,y2-y1)))
    return boxes

def detect_haar(img_bgr, scale=1.05, neigh=5, minsize=(60,60)):
    gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
    return list(haar.detectMultiScale(gray, scaleFactor=scale, minNeighbors=neigh, minSize=minsize))

def predict_roi(roi_bgr):
    roi_rgb = cv2.cvtColor(roi_bgr, cv2.COLOR_BGR2RGB)
    r = cv2.resize(roi_rgb, IMG_SIZE)
    inp = (r.astype("float32")/255.0)[None,...]
    device = '/GPU:0' if tf.config.list_physical_devices('GPU') else '/CPU:0'
    with tf.device(device):
        prob = float(model.predict(inp, verbose=0)[0][0])
    # interpret (0->with_mask, 1->without_mask)
    if prob>0.5:
        return idx2label[1], prob
    else:
        return idx2label[0], 1.0-prob

def robust_predict(image_path, show=True):
    img = cv2.imread(image_path)
    if img is None: raise FileNotFoundError(image_path)
    H,W = img.shape[:2]

    # 1) DNN SSD
    if dnn_net is not None:
        boxes = detect_dnn(img, dnn_net, conf_thresh=0.4)
        method = "dnn_ssd"
        if boxes:
            print("Detected by DNN SSD:", len(boxes))
    else:
        boxes=[]; method=None

    # 2) Haar if DNN failed
    if not boxes:
        boxes = detect_haar(img, scale=1.05, neigh=5, minsize=(40,40))
        method = method or "haar"
        if boxes:
            print("Detected by Haar:", len(boxes))

    # 3) Try upscaling if still none
    if not boxes:
        for s in (1.5, 2.0):
            up = cv2.resize(img, (int(W*s), int(H*s)))
            boxes_up = detect_haar(up, scale=1.05, neigh=3, minsize=(30,30))
            if boxes_up:
                boxes = [(int(x/s), int(y/s), int(w/s), int(h/s)) for (x,y,w,h) in boxes_up]
                method = f"haar_up_{s}"
                print("Detected after upscaling x", s)
                break

    # 4) Sliding-window fallback if still none (will be slow)
    if not boxes:
        print("[fallback] sliding-window scanning (slow).")
        candidates=[]
        scales = [1.0, 0.8, 0.6]
        for s in scales:
            win = int(min(H,W)*s)
            if win<30: continue
            step = max(8, win//3)
            for y in range(0, H-win+1, step):
                for x in range(0, W-win+1, step):
                    crop = img[y:y+win, x:x+win]
                    label, conf = predict_roi(crop)
                    candidates.append((x,y,win,win,label,conf))
        # sort by confidence desc and pick top K
        candidates = sorted(candidates, key=lambda x: x[5], reverse=True)[:6]
        boxes = [(c[0],c[1],c[2],c[3]) for c in candidates]
        method = "sliding_window"
        preds = [{"box":[int(x),int(y),int(w),int(h)], "label":label, "confidence":float(conf)} for (x,y,w,h,label,conf) in candidates]
        # show
        if show:
            for p in preds:
                x,y,w,h = p["box"]; crop = img[y:y+h, x:x+w];
                plt.figure(figsize=(3,3)); plt.axis('off'); plt.title(f"{p['label']} {p['confidence']:.2f}"); plt.imshow(cv2.cvtColor(crop, cv2.COLOR_BGR2RGB)); plt.show()
            plt.figure(figsize=(6,8)); plt.axis('off'); plt.title("Top sliding-window candidates"); plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)); plt.show()
        return {"status":"ok","method":method,"predictions":preds}

    # If we have boxes from detectors:
    preds=[]
    for (x,y,w,h) in boxes:
        pad = int(0.12*max(w,h))
        x0,y0 = max(0,x-pad), max(0,y-pad)
        x1,y1 = min(W,x+w+pad), min(H,y+h+pad)
        crop = img[y0:y1, x0:x1]
        label, conf = predict_roi(crop)
        preds.append({"box":[int(x0),int(y0),int(x1-x0),int(y1-y0)], "label":label, "confidence":float(conf)})
        # draw
        color=(0,255,0) if "mask" in label.lower() else (255,0,0)
        cv2.rectangle(img, (x0,y0), (x1,y1), color, 2)
        cv2.putText(img, f"{label} {conf:.2f}", (x0,y0-10), cv2.FONT_HERSHEY_SIMPLEX, 0.7, color, 2)

    if show:
        plt.figure(figsize=(8,6)); plt.axis('off'); plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)); plt.show()
    return {"status":"ok","method":method,"predictions":preds}

# Example usage:
example = "/kaggle/input/datasetaa/red-apple-fruit.jpg"
print("Testing:", example)
out = robust_predict(example, show=True)
print(out)
```
